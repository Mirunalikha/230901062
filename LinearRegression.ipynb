```python
import matplotlib.pyplot as plt
from scipy import stats

x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,89,79,11,111,35,57,88,44,94,123,89]

slope,intercept,r,p,std_err=stats.linregress(x,y)

def mf(x):
    return slope * x + intercept

mm=list(map(mf,x))

plt.scatter(x,y)
plt.plot(x,mm)
plt.show()
```
